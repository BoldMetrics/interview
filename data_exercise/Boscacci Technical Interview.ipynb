```python
import pandas as pd
import json
```

```python
with open('astronaut_product.json', 'r') as jfile:
    astronaut_product = json.load(jfile)
```

```python
products = list()
```

```python
product_title = "Astronaut Suit"

description = ("Enhances space-walking experience, keeps oxygen inside suit."
               "Our most popular color is a flashy white with NASA patches. "
               "Helmet and space-walking chair sold separately.")
```

```python
products_from_data = astronaut_product['products']
products_dict = products_from_data[0]
products_dict
```

```python
products_dict.keys()
```

```python
df_variants = pd.DataFrame(products_dict['variants'])
```

```python
df_variants_less = df_variants['id title price inventory_quantity option1'.split()].copy()
df_variants_less.rename(columns={'option1':'color'},inplace=True)
# df_variants_less
```

```python
df_variants_less['combined_titles'] = product_title + df_variants_less['title']
```

```python
df_variants_less['description'] = description
# df_variants_less
```

```python
df_variants_less['price'] = df_variants_less.price.astype(float)
# df_variants_less
```

```python
df_variants_less['potential_sales'] = df_variants_less.inventory_quantity.astype(float) * df_variants_less.price
df_variants_less
```

```python
potential_sales_total = df_variants_less.potential_sales.sum()
potential_sales_total
```

```python
df_variants_less['in_stock'] = df_variants_less.inventory_quantity > 0
df_variants_less
```
